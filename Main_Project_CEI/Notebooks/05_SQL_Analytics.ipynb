# Project Sentinel – SQL Analytics

## Objective

This notebook performs SQL-based analytics on the Gold layer to derive business insights from transactional data. It demonstrates how Delta tables can be queried using SQL for reporting and decision-making.

### Features

- Revenue Analysis
- Customer Analytics

### Execution Flow

Input:
- Gold Delta Tables

Processing:
- SQL queries
- Business analysis
- KPI reporting

Output:
- Analytical insights

```python
BASE_PATH = "/Volumes/workspace/default/project_sentinel"

GOLD_PATH = f"{BASE_PATH}/gold"
```

```python
spark.read.format("delta") \
    .load(f"{GOLD_PATH}/total_revenue") \
    .createOrReplaceTempView("total_revenue")
```

```python
spark.read.format("delta") \
    .load(f"{GOLD_PATH}/top_products") \
    .createOrReplaceTempView("top_products")
```

```python
spark.read.format("delta") \
    .load(f"{GOLD_PATH}/top_customers") \
    .createOrReplaceTempView("top_customers")
```

```python
spark.read.format("delta") \
    .load(f"{GOLD_PATH}/payment_distribution") \
    .createOrReplaceTempView("payment_distribution")
```

```python
display(
    spark.sql("""
    SELECT *
    FROM total_revenue
    """)
)
```

```python
display(
    spark.sql("""
    SELECT *
    FROM top_products
    ORDER BY Total_Quantity DESC
    """)
)
```

```python
display(
    spark.sql("""
    SELECT *
    FROM top_customers
    ORDER BY Revenue DESC
    """)
)
```

```python
display(
    spark.sql("""
    SELECT *
    FROM payment_distribution
    ORDER BY count DESC
    """)
)
```
