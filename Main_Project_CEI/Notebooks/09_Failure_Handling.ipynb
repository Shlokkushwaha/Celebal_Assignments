```python
BASE_PATH = "/Volumes/workspace/default/project_sentinel"

BRONZE_PATH = f"{BASE_PATH}/bronze"

bronze_df = spark.read.format("delta").load(BRONZE_PATH)
```

```python
bronze_df.printSchema()
```

```python
bad_records = bronze_df.filter(
    col("_rescued_data").isNotNull()
)

display(bad_records)
```

```python
total_records = bronze_df.count()

good_records = bronze_df.filter(
    col("_rescued_data").isNull()
).count()

bad_records = bronze_df.filter(
    col("_rescued_data").isNotNull()
).count()

print("="*50)
print("PROJECT SENTINEL DATA QUALITY REPORT")
print("="*50)

print(f"Total Records      : {total_records}")
print(f"Good Records       : {good_records}")
print(f"Rescued Records    : {bad_records}")

print("="*50)
```

```python
from pyspark.sql.types import *

schema = StructType([
    StructField("transaction_id", StringType(), True),
    StructField("transactional_date", StringType(), True),
    StructField("product_id", StringType(), True),
    StructField("customer_id", StringType(), True),
    StructField("payment", StringType(), True),
    StructField("credit_card", StringType(), True),
    StructField("loyalty_card", StringType(), True),
    StructField("cost", StringType(), True),
    StructField("quantity", StringType(), True),
    StructField("price", StringType(), True)
])

bad_data = [
    ("ABC", "INVALID_DATE", "P100", None, "UPI", "Yes", "No", "abc", "xyz", "150")
]

bad_df = spark.createDataFrame(bad_data, schema)

display(bad_df)
```
