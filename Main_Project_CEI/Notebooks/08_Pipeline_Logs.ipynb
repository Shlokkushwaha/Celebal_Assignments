# Project Sentinel – Pipeline Logging

## Objective

This notebook generates execution logs for each stage of the data pipeline, providing operational visibility into processing status, record counts, and execution timestamps.

### Features
- Pipeline Audit Logs
- Execution Status
- Processing Metrics
- Timestamp Logging
- Operational Monitoring
- Pipeline Traceability

### Execution Flow

Purpose:
Generate execution logs for every pipeline stage.

Logs:
- Layer Name
- Record Count
- Status
- Timestamp

```python
from pyspark.sql import Row
from pyspark.sql.functions import current_timestamp
```

```python
BASE_PATH = "/Volumes/workspace/default/project_sentinel"

LOG_PATH = f"{BASE_PATH}/logs/pipeline_logs"
```

```python
BRONZE_PATH = f"{BASE_PATH}/bronze"
SILVER_PATH = f"{BASE_PATH}/silver"
GOLD_PATH = f"{BASE_PATH}/gold"

bronze_df = spark.read.format("delta").load(BRONZE_PATH)
silver_df = spark.read.format("delta").load(SILVER_PATH)

top_products = spark.read.format("delta").load(f"{GOLD_PATH}/top_products")
top_customers = spark.read.format("delta").load(f"{GOLD_PATH}/top_customers")
payment_distribution = spark.read.format("delta").load(f"{GOLD_PATH}/payment_distribution")
```

```python
logs = [
    Row(layer="Bronze",
        records=bronze_df.count(),
        status="SUCCESS"),

    Row(layer="Silver",
        records=silver_df.count(),
        status="SUCCESS"),

    Row(layer="Gold_Top_Products",
        records=top_products.count(),
        status="SUCCESS"),

    Row(layer="Gold_Top_Customers",
        records=top_customers.count(),
        status="SUCCESS"),

    Row(layer="Gold_Payment_Distribution",
        records=payment_distribution.count(),
        status="SUCCESS")
]

log_df = spark.createDataFrame(logs)

log_df = log_df.withColumn(
    "execution_time",
    current_timestamp()
)
```

```python
log_df.write \
      .format("delta") \
      .mode("overwrite") \
      .save(LOG_PATH)
```

```python
logs = spark.read.format("delta").load(LOG_PATH)

display(logs)
```

```python
print("="*60)
print("PROJECT SENTINEL PIPELINE EXECUTION REPORT")
print("="*60)

display(logs)
```
