# Project Sentinel – Pipeline Monitoring

## Objective

This notebook monitors the health and performance of the data pipeline by tracking record counts, ingestion timestamps, pipeline status, and storage metrics across Bronze, Silver, and Gold layers.

### Features
- Pipeline Health Monitoring
- Record Count Validation
- Data Freshness Tracking
- Layer-wise Monitoring
- Operational Dashboard

### Execution Flow

Purpose:
Monitor the overall pipeline health.

Monitors:
- Bronze Records
- Silver Records
- Gold Records
- Latest Ingestion Time
- Pipeline Status

```python
BASE_PATH = "/Volumes/workspace/default/project_sentinel"

BRONZE_PATH = f"{BASE_PATH}/bronze"
SILVER_PATH = f"{BASE_PATH}/silver"
GOLD_PATH = f"{BASE_PATH}/gold"

RAW_PATH = f"{BASE_PATH}/raw"

CHECKPOINT_PATH = f"{BASE_PATH}/checkpoints/bronze"
```

```python
bronze_df = spark.read.format("delta").load(BRONZE_PATH)

silver_df = spark.read.format("delta").load(SILVER_PATH)

total_revenue = spark.read.format("delta").load(f"{GOLD_PATH}/total_revenue")

top_products = spark.read.format("delta").load(f"{GOLD_PATH}/top_products")

top_customers = spark.read.format("delta").load(f"{GOLD_PATH}/top_customers")

payment_distribution = spark.read.format("delta").load(f"{GOLD_PATH}/payment_distribution")
```

```python
print("="*50)
print("PROJECT SENTINEL MONITORING DASHBOARD")
print("="*50)

print(f"Bronze Records : {bronze_df.count()}")

print(f"Silver Records : {silver_df.count()}")

print(f"Top Products : {top_products.count()}")

print(f"Top Customers : {top_customers.count()}")

print(f"Payment Methods : {payment_distribution.count()}")

print("="*50)
```

```python
from pyspark.sql.functions import max

bronze_df.select(
    max("ingestion_time").alias("Latest_Ingestion")
).show(truncate=False)
```
